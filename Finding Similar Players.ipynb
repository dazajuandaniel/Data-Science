```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_excel('datasets/CompleteDataset.xlsx',
                       sheet_name = 'CompleteDataset (2)',
                       index_col =0)
```

```python
dataset.head(3)
```

```python
columns_to_compare = ['ID','Name','Age',"Acceleration","Aggression","Agility","Balance",'Ball control',
                      'Crossing','Dribbling',"Finishing",'Long passing','Long shots','Short passing','Shot power',
                     'Sliding tackle','Sprint speed','Stamina','Strength','Vision']
```

```python
data = dataset.loc[:,columns_to_compare].copy()
data.head(1)
```

```python
player_name = 'Iniesta'
no_attributes = 5
by_index = data[data['Name'] == player_name].index[0]
```

```python
cols = data[data['Name'] == player_name].loc[:,data.columns!='Name'].sort_values(
    by = by_index,axis=1,ascending =False).columns[:no_attributes]
```

```python
cols = list(cols)
cols
```

```python
comparison_vec = data[data['Name'] == player_name][cols]
ref = comparison_vec.iloc[0,:]
ref
```

```python
data_to_compare = dataset.loc[:,cols]
data_to_compare.head(1)
```

```python
dataset['similarity'] = abs(data_to_compare.sub(ref, axis=1)).sum(axis=1).to_frame()
```

```python
dataset.sort_values(by = 'similarity')
```
